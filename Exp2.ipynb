```python
Q2="""AIM:Palindrome check and diggit occurrence count:develop a python program to check whether a given string is palindrome or not and also count the number of occurrence of each letter in the input string """
print(Q2)

input_str = input("Enter a string: ")
if input_str == input_str[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")

for ch in set(input_str):
    count = input_str.count(ch)
    print(f"'{ch}' appears {count} times")
result="Result:This program executed successfully"
print(result)
```
